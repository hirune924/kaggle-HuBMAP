```python
import tifffile as tiff
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
import os
```

```python
for h in range(10,20):
    for w in range(20):
        mask = np.asarray(Image.open('data/size_1024/w{}-h{}-{}_mask.png'.format(w,h,'cb2d976f4')))
        if np.max(mask)==1:
            img = np.asarray(Image.open('data/size_1024/w{}-h{}-{}.png'.format(w,h,'cb2d976f4')))
            print('{}:{}'.format(h,w))
            break
    else:
        continue
    break
plt.imshow(mask)
print(np.unique(mask))
print(mask.shape)
plt.figure(figsize=(15, 15))
plt.imshow(img)
plt.imshow(mask, alpha=0.5)
plt.title("Image + Mask", fontsize=16);
plt.axis("off");
```
